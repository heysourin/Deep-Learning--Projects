# **Activation function**

## **What is an activation function?**
An application of the activation function to a neuron's (or layer of neurons') output affects outputs. We employ activation functions because they enable neural networks, which typically have two or more hidden layers, to map nonlinear functions if the activation function itself is nonlinear. Without activation function neural network will behave as linear function, which is unable to map non-linear data.

## **Step Activation function**

  image.png


```python
def step_function(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
step_function(-1), step_function(0), step_function(2)
```

```python
# Sigmoid function
import numpy as np
def sigmoid_function(x):
  z = (1/(1 + np.exp(-x)))
  return z
```

## **Tanh: Hyperbolic Tangent**
- When our problem domain requires values from -1 to +1.
- Hyperbolic
-
  image.png
- image.png
- Gradient of tanh also has vanishing gradient problem:
image.png

```python
import numpy as np
def tanh_function(x):
  z = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
  return z
```

```python
# another way of writing the same
def tanh_function(x):
    z = (2/(1 + np.exp(-2*x))) -1
    return z
```

```python
tanh_function(0.5), tanh_function(-1)
```

## **ReLU Activation Function**
image.png

```python
def relu_function(x):
    if x<0:
        return 0
    else:
        return x
```

## **Leaky ReLU**
image.png

```python
def leaky_relu_function(x):
    if x<0:
        return 0.01*x
    else:
        return x
```

```python
leaky_relu_function(7), leaky_relu_function(-7)
```

# **Choosing right activation function**

*   Sigmoid functions and their combinations generally work better in the case of classifiers.
* Sigmoids and tanh functions are sometimes avoided due to the vanishing gradient problem.
* ReLU function is a general activation function and is used in most cases these days.
* If we encounter a case of dead neurons in our networks the leaky ReLU function is the best choice.
* Always keep in mind that ReLU function should only be used in the hidden layers.
* As a rule of thumb, you can begin with using ReLU function and then move over to other activation functions in case ReLU doesn’t provide with optimum results
